```python
import pandas as pd
import numpy as np
```

```python
path='../datasets/LUN/fulltrain.csv'
dataset=pd.read_csv(path,header=None)
```

```python
dataset.head(5)
```

```python
cols=('category','content')
```

```python
dataset.columns=cols
dataset.head()
type(dataset['category'][0])
```

```python
# for category !=4, define the fake label
def set_label(category):
    if category!=4:
        return 'fake'
    else:
        return 'real'
dataset['labels']=dataset['category'].apply(lambda x :set_label(x))
```
