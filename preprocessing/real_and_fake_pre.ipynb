```python
raw_data_path = 'real_and_fake.csv'
destination_folder = '../datasets/real_and_fake_new/'
train_test_ratio = 0.10
train_valid_ratio = 0.80
# trim the words in the news
first_n_words = 512
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('../model_files/bert-base-uncased')


def trim_string(x):
    x = x.split(maxsplit=first_n_words)
    x = ' '.join(x[:first_n_words])
    return x


def get_attention_mask(x):
    return tokenizer(x, add_special_tokens=True, return_attention_mask=True, return_tensors='pt')['attention_mask'].tolist()
```

```python
  # Read raw data
df_raw = pd.read_csv(raw_data_path)

# Prepare columns
df_raw['label'] = (df_raw['label'] == 'FAKE').astype('int')
df_raw = df_raw.reindex(columns=[ 'title', 'text', 'label'])

# Drop rows with empty text
df_raw.drop(df_raw[df_raw.text.str.len() < 5].index, inplace=True)

# Trim text and titletext to first_n_words
df_raw['label']=df_raw['label']
df_raw['text'] = df_raw['text']
df_raw.head()
```
